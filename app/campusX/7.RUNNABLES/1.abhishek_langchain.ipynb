```python
import random

class AbhishekLLM:
  def __init__(self):
    print("llm created")
  def predict(self,prompt):
    response_list = [
        "Delhi is the capital of india",
        "ipl is a cricket league",
        "ai means artificial inteligence"
    ]

    return {'response':random.choice(response_list)}


a = AbhishekLLM()
```

```python
class AbhishekPromptTemplate:
  def __init__(self,template,input_variables):
    self.template = template
    self.input_variables = input_variables
  def format(self,input_dict):
    return self.template.format(**input_dict)


b = AbhishekPromptTemplate(
    template = "what is the capital of {country} in {length}",
    input_variables = ["country"]
)


b.format(input_dict = {"country":"india","length":"short"})
```

```python
prompt = AbhishekPromptTemplate(
    template = "what is the capital of {country} in {length}",
    input_variables = ["country"]
)


final = prompt.format(input_dict = {"country":"india","length":"short"})


ki = a.predict(final)


print(ki)
```

```python
class AbhishekLLMChain:
  def __init__(self,llm,prompt):
    self.llm = llm
    self.prompt = prompt
  def run(self, input_dict):
    prompt_text = self.prompt.format(input_dict)
    return self.llm.predict(prompt_text)

```

```python
testa = AbhishekLLMChain(
    llm = a,
    prompt = prompt
)

print(testa.run( {"country":"india","length":"short"}))
```
